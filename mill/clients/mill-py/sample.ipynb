Import Mill

```python
from millclient import *
```

Create channel and client
MillService should be runing on locahost:9099 TLS must be disabled

```python
channel = Channel(host = "localhost", port = 9099)
client = create_client(channel=channel)
```

Create query and execute to pandas
In notebooks only async version of methods can be used

```python
q1 = await client.sql_query(sql = "SELECT * FROM `airlines`.`passenger` ", fetch_size = 10).to_pandas_async()
display(q1.head(10))
```

```python
q = client.sql_query(sql = "SELECT c.`id` AS `city_id`, p.`id` as `passenger_id`, c.*, p.`domicile_city_id` as `passenger_city_id` FROM `airlines`.`cities` c inner join `airlines`.`passenger` p ON c.`id` = p.`domicile_city_id`" , fetch_size = 10)
pd = await q.to_pandas_async()
display(pd.head(10))
```
